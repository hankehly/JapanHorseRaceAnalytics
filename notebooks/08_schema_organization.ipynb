```python
from pyspark.sql import SparkSession

from JapanHorseRaceAnalytics.utilities.base import get_base_dir


warehouse_dir = f"{get_base_dir()}/spark-warehouse"
postgres_driver_path = f"{get_base_dir()}/jars/postgresql-42.7.1.jar"

spark = (
    SparkSession.builder.appName("20240211_competitors")
    .config("spark.driver.memory", "21g")
    .config("spark.driver.maxResultSize", "5g")
    .config("spark.sql.warehouse.dir", warehouse_dir)
    .config("spark.jars", postgres_driver_path)
    .config("spark.executor.extraClassPath", postgres_driver_path)
    .config("spark.driver.extraClassPath", postgres_driver_path)
    .enableHiveSupport()
    .getOrCreate()
)
```

```python
names = spark.read.table("jhra_curated.features_20240217_v1").schema.names

names_1 = [name for name in names if "_事前_" not in name]
names_1 = [name.replace("_実績", "") if "_実績" in name else name for name in names_1]

names_2 = [name for name in names if "_実績_" not in name]
names_2 = [name.replace("_事前", "") if "_事前" in name else name for name in names_2]

sorted(names_1) == sorted(names_2)
```

```python
# Get all the meta_ columns
meta_columns = sorted([name for name in names if "meta_" in name])
# Get all the num_ columns
num_columns = sorted([name for name in names if "num_" in name])
# Get all the cat_ columns
cat_columns = sorted([name for name in names if "cat_" in name])
# Combine the columns
columns = meta_columns + num_columns + cat_columns
columns
```
